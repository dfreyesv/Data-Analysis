```python
import pandas as pd
import os
```

```python
csv_dir = "C:\\Users\\dfrv9\\Documents\\NBA\\csv files"
csv_files_list = os.listdir(csv_dir)
csv_files_list = [f for f in csv_files_list if f.endswith(".csv") and "games" in f]
```

```python
stats = []
for csv in csv_files_list:
    read_path = os.path.join(csv_dir,csv)
    df = pd.read_csv(read_path)
    stats.append(df)
df = pd.concat(stats)
```

```python
df = df.sort_values("date")
df = df.reset_index(drop=True)
```

```python
def add_target(team):
    team["target"] = team["won"].shift(-1)
    
    return team

df = df.groupby("team", group_keys=False).apply(add_target)
```

```python
df.head()
```

```python
df["target"][pd.isnull(df["target"])] = 2
```

```python
df["target"] = df["target"].astype(int, errors="ignore")
```

```python
df
```

```python
nulls = pd.isnull(df)
nulls = nulls.sum()
nulls = nulls[nulls>0]
```

```python
df = df.drop(nulls.index, axis=1)
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rr = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, n_features_to_select=30, direction="forward", cv=split)
```

```python
removed_colums = ["season", "date", "won", "target", "Team", "Team_opp"]
selected_colums = df.columns[~df.columns.isin(removed_colums)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[selected_colums] = scaler.fit_transform(df[selected_colums])
```

```python
sfs.fit(df[selected_colums], df["target"])
```

```python
predictors = list(selected_colums[sfs.get_support()])
```

```python
predictors
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []
    
    seasons = sorted(data["season"].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        
        train = data[data["season"] < season]
        test = data[data["season"] == season]
        
        model.fit(train[predictors], train["target"])
        
        preds = model.predict(test[predictors])
        
        preds = pd.Series(preds, index=test.index)
        
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df, rr, predictors)
```

```python
predictions
```

```python
from sklearn.metrics  import accuracy_score

accuracy_score(predictions["actual"], predictions["prediction"])
```
